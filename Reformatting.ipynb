```python
import chess.pgn
import os
from threading import Thread
import time
```

```python
def reformat_games( file_names, new_dir, thread_idx ):
    file_name_idx = 0
    #Iterate over all file names
    while len(file_names) > 0: # Go through files until there's no more
        fname = file_names.pop()
        pgn_fh = open(fname) # Open PGN
        # Iterate over all games in file
        while True:
            game = chess.pgn.read_game(pgn_fh) # Read PGN singularly, iterates through it until no more
            if not game: # Nothing detected 
                break
                
            with open(f'{new_dir}/{thread_idx}_{file_name_idx}.pgn', "a") as file:
                  print(game, file=file, end="\n\n")
            file_name_idx += 1
            if file_name_idx % 1000 == 0: # Every 1000 PGN's prints out how much it's done
                print( 'Thread {} wrote {} train pngs'.format( thread_idx, file_name_idx ) )
        pgn_fh.close()
    print(f"Done {thread_idx} {len(file_names)}")

```

```python
ccrl_dir = 'cclr/train'
all_file_names = os.listdir( ccrl_dir )
for x in all_file_names:
    if not x.endswith(".pgn"):
        all_file_names.pop(all_file_names.index(x))
        print("removed smth")
for i in range( len( all_file_names ) ):
    all_file_names[ i ] = os.path.join( ccrl_dir, all_file_names[ i ] ) 
print(len(all_file_names))
```

```python
reformat_dir = 'cclr/reformat'

#launch some threads
threads = []
num_threads = 50
for i in range( num_threads ):
    files_per_thread = int( len( all_file_names ) / num_threads )
    threads.append( Thread( target=reformat_games,
        args=( all_file_names[ i * files_per_thread : (i + 1) * files_per_thread ],
            reformat_dir, i ) ) )
    threads[ i ].start()
    time.sleep( 0.0001 )

for i in range( num_threads ):
    threads[ i ].join()

```

```python
import os

def get_size(start_path = '.'):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(start_path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            # skip if it is symbolic link
            if not os.path.islink(fp):
                total_size += os.path.getsize(fp)

    return total_size

print(get_size("rere"), 'bytes')
```
